```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
start_time = time.time()

# Preparation of the differents data sets for training and testing
X_train = train_df.drop("class", axis=1)
y_train = pd.DataFrame(data=train_df["class"], columns=["class"])
X_test = test_df.drop("class", axis=1)
y_test = pd.DataFrame(data=test_df["class"], columns=["class"])

# Building the model
# Passing trough a lot of different combinaison for the two metaparameters
all_accur = []  # All the accuracies computed of each combinaison of metaparameters
means_accur = []
for i in range(15,40):  # n_estimators
    accur = []
    for j in range(2,5): # max_leaf_nodes = 2**j
        mean_test = []
        for k in range(2):  # numbers trials fo each combinaison of metaparameters
            RFC = RandomForestClassifier(n_estimators=i, max_leaf_nodes=2**j)
            RFC.fit(X_train, y_train)
            RFC_pred = RFC.predict(X_test)
            score = accuracy_score(y_test, RFC_pred)
            mean_test.append(score)
        accur.append(np.mean(mean_test))  # Computation of the mean accuracy for the 20 trials
    all_accur.append(accur)
    means_accur.append(np.mean(accur))

# Searching for the best combinaison of metaparameters
# Function found on internet : https://thispointer.com/python-find-indexes-of-an-element-in-pandas-dataframe/
def getIndexes(dfObj, value):
    ''' Get index positions of value in dataframe i.e. dfObj.'''
    listOfPos = list()
    # Get bool dataframe with True at positions where the given value exists
    result = dfObj.isin([value])
    # Get list of columns that contains the value
    seriesObj = result.any()
    columnNames = list(seriesObj[seriesObj == True].index)
    # Iterate over list of columns and fetch the rows indexes where value exists
    for col in columnNames:
        rows = list(result[col][result[col] == True].index)
        for row in rows:
            listOfPos.append((row, col))
    # Return a list of tuples indicating the positions of value in the dataframe
    return listOfPos

acc = np.array(all_accur) 
t1 = pd.DataFrame(acc, columns =["n_estimators", "1", "2"]) # Creating a dataframe with all the possible combinaisons of metaparameters

max_accur = t1.max().max() # Finding the highest accuracy in the dataframe
indexes = getIndexes(t1, max_accur) # Finding the position of the highest accuracy
index = indexes[0]

N_estimators = index[0]+15  # Best number of estimators 
Max_leaf_nodes = 2**(int(index[1])+2) # Best number of maximum leaf nodes

my_best_forest = RandomForestClassifier(n_estimators=N_estimators, max_leaf_nodes=Max_leaf_nodes)

print("--- %s seconds ---" % (time.time() - start_time))
```
